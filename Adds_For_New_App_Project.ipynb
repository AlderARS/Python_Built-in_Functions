# Limpeza e separa√ß√£o de dados

### Verificar se h√° linhas duplicadas
√â necess√°rio verificar se existem duplicatas em nosso conjuntos de dados, afinal, n√£o √© interessante contrar o mesmo dado duas ou mais vezes. 

Para fazer essa verifica√ß√£o criarei uma fun√ß√£o para analisar o nome dos aplicativos dentro de uma lista de listas

```python
### Fun√ß√£o para verificar nomes duplicados dentro dos conjuntos de dados ###
def verify_duplicates(dataset):
    duplicate_apps = []
    unique_apps = []

    for app in dataset:
        name = app[0]
        if name in unique_apps:
            duplicate_apps.append(name)
        else:
            unique_apps.append(name)
    
    print('N√∫mero de aplicativos duplicados:', len(duplicate_apps))
    print('\n')
    print('Examples de aplicativos duplicados:', duplicate_apps[:15])
    
verify_duplicates(ios) # 0 nomes duplicados
print('\n')
verify_duplicates(android) # 1.181 nomes duplicados
```

```python
### An√°lisando duplicatas ###
for app in android:
    name = app[0]
    if name == 'Quick PDF Scanner + OCR FREE':
        print(app)
```

### Separando duplicatas
Algumas duplicatas s√£o exatamente iguais por√©m v√™-se alguma diferen√ßa na quarta coluna que corresponde ao numero de avalia√ß√µes.

N√£o deve-se excluir linhas aleat√≥riamente pois existe a chance do numero de avalia√ß√µes mais alto pode ser o mais atual. Quero manter a duplicata com o maior numero de avalia√ß√µes e descartar as outras.

```python
### Pegando apenas as duplicatas com mais avalia√ß√µes ###
reviews_max = {}

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

```python
print('Comprimento esperado:', len(android) - 1181)
print('Comprimento atual:', len(reviews_max)) # 9.659 linhas de comprimento
```

### Separa√ß√£o: Apenas valores √∫nicos

```python
### Dicion√°rio reviews_max para criar uma lista sem duplicatas ###
android_clean = []
already_added = []

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

```python
### Verificando comprimento da nova lista ###
explore_data(android_clean, 0, 1, True) # 9.659 linhas de comprimento.
```

Sabe-se que cada caractere de uma string corresponde a um √≠ndice(n√∫mero) ent√£o para facilitar a separa√ß√£o deve-se separar os caracteres que s√£o comuns no idioma ingl√™s das que n√£o s√£o. 

Tamb√©m existem nomes de aplicativos com alguns caracteres diferentes, para n√£o perder dados colocaremos um limite de 3 caracteres diferentes do normal por nome.

**Segundo a ASCII os √≠ndices dos caracteres que desejamos est√£o entre os √≠ndices 0 e 127.**

```python
### Fun√ß√£o para verificar √≠ndices dos caracteres com limite de 3 Falses ###
def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True

print(is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))
print(is_english('Instachat üòú'))
```

```python
### Pegamos listas sem arquivos duplicados e verificaremos se is_english ###
android_english = []
ios_english = []

for app in android_clean:
    name = app[0]
    if is_english(name):
        android_english.append(app)
        
for app in ios:
    name = app[1]
    if is_english(name):
        ios_english.append(app)
        
explore_data(android_english, 0, 1, True)
print('\n')
explore_data(ios_english, 0, 1, True)
```

# An√°lise: Visualizando a Superf√≠cie
nosso objetivo √© determinar os tipos de aplicativos que provavelmente atrair√£o mais usu√°rios, porque o n√∫mero de pessoas que usam nossos aplicativos afeta nossa receita.

Como nosso objetivo final √© adicionar o aplicativo ao Google Play e √† App Store, precisamos encontrar perfis de aplicativos que sejam bem-sucedidos em ambos os mercados.

```python
### Fun√ß√£o para encontrar a frequencia de uma coluna numa lista de listas em porcentagem ###
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        percentage = round(percentage, 2)
        table_percentages[key] = percentage 
    
    return table_percentages

### Fun√ß√£o para ordenar a fun√ß√£o acima ###
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

## Apps da App Store com mais usu√°rios
O conjunto de dados da App Store n√£o tem uma coluna especifica para dizer a quantidade de instala√ß√µes por aplicativo. Levando em considera√ß√£o que para se avaliar um aplicativo deve-se t√™-lo usado/instalado antes, irei usar o total de avalia√ß√µes para nos dar uma base sobre a m√©dia de instala√ß√µes por g√™nero.

```python
### Encontrando a m√©dia de instala√ß√µes por genero na App Store ###
genres_ios = freq_table(ios_final, -5)
for genre in genres_ios:
    total_ratings = 0
    len_genre = 0
    for app in ios_final:
        genre_app = app[-5]
        if genre_app == genre:            
            n_ratings = float(app[5])
            total_ratings += n_ratings
            len_genre += 1
    avg_n_ratings = total_ratings / len_genre
    avg_n_ratings = round(avg_n_ratings, 2)
    print(genre, ':', avg_n_ratings)
```

## Apps da Google Play com mais usu√°rios

```python
### Encontrando a m√©dia de instala√ß√µes por genero no Google Play ###
category_android = freq_table(android_final, 1)
for category in category_android:
    total_installs = 0
    len_category = 0
    for app in android_final:
        category_app = app[1]
        if category_app == category:
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            n_installs = float(n_installs)
            total_installs += n_installs
            len_category += 1
    avg_n_ratings = total_installs / len_genre
    avg_n_ratings = round(avg_n_ratings, 2)
    print(category, ':', avg_n_ratings)
```
